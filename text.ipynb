```python
import numpy as np

# 加载.npy文件
data = np.load('/media/a/软件/BaiduNetdiskDownload/SEDCNN4(Bamboo)/SEDCNN4_WAXD_surpervised-Bamboo/npy_img/dot1_masked_and_norm.npy')
```

```python
import matplotlib.pyplot as plt

# 选择要显示的图像索引，例如第一个图像
index = 0

# 显示第 index 个图像
plt.imshow(data[index], cmap='gray')  # 使用灰度图显示
plt.colorbar()
plt.show()
```

```python
# 显示所有图像
fig, axes = plt.subplots(5, 5)  # 创建一个5x5的子图网格
axes = axes.flatten()  # 将二维的轴数组展平为一维

for i in range(25):  # 假设我们只想显示前25个图像
    axes[i].imshow(data[i], cmap='gray')  # 使用灰度图显示
    axes[i].axis('off')  # 隐藏坐标轴
plt.show()
```
